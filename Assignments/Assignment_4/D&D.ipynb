```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def roll_n_side_m_times(num_sides, num_rolls):
    rolls = np.random.randint(1, num_sides, num_rolls)
    plt.hist(rolls, range(1, num_sides + 1))
    return rolls.mean()
```

```python
roll_n_side_m_times(20, 1000000)
```

```python
def max_roll_n_side_m_times(num_sides, num_rolls):
    rolls = np.random.randint(1, num_sides, num_rolls)
    return rolls.max()
```

```python
rolls = [max_roll_n_side_m_times(20, 2) for _ in range(1000000)]
plt.hist(rolls, range(1, 20 + 1))
```
