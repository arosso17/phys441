```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
r = 1
m = 1
I = (2*m*r*r)/5

dt = 0.01

pos = np.array([0.0, 50.0])
vel = np.array([0.0, 0.0])
acc = np.array([0.0, 0.0])
g = -9.81
t = 0
vels = [vel.copy()]
poss = [pos.copy()]
accs = [acc.copy()]
times = [0]
while t < 10:
    acc = np.array([0.0, 0.0])
    acc[1] += g
    if pos[1] < 0 + r:
        # print(f"Hit a wall at {t} with pos {pos}")
        # print(f"\tCurrent vel: {vel}")
        vel[1] *= -1
        # print(f"\tChanges it to: {vel}")
    t += dt
    pos += vel * dt + 1/2 * acc * dt * dt
    vel += (accs[-1] + acc) / 2 * dt
    poss.append(pos.copy())
    vels.append(vel.copy())
    accs.append(acc.copy())
    times.append(t)
    
fig, ax = plt.subplots(1, 3)
ax[0].plot(times, [p[0] for p in poss])
ax[0].set_xlabel("t")
ax[0].set_ylabel("x")
ax[1].plot(times, [p[1] for p in poss])
ax[1].set_xlabel("t")
ax[1].set_ylabel("y")
ax[2].plot([p[0] for p in poss], [p[1] for p in poss])
ax[2].set_xlabel("x")
ax[2].set_ylabel("y")
plt.show()
```

```python
r = 1
m = 1
I = (2*m*r*r)/5

kty = 0.5
ktx = 0
kr = 0.9

dt = 0.1
g = -9.1

pos = np.array([0.0, 25])
vel = np.array([1.0, -20.0])
acc = np.array([0.0, 0])
rot_vel = 0
angle = 0

t = 0

vels = [vel.copy()]
poss = [pos.copy()]
accs = [acc.copy()]
angles = [angle]
spins = [rot_vel]
times = [t]
while t < 10:
    acc = np.array([0.0, 0.0])
    acc[1] += g
    if pos[1] > 25 or pos[1] < 0:
        if pos[1] < 0:
            pos[1] = 0
        if pos[1] > 25:
            pos[1] = 25
        print(f"Hit a wall at {t} with pos {pos}")
        print(f"\tCurrent vel: {vel}")
        e_tx = np.sign(vel[0]) * 1/2 * m * vel[0] ** 2
        e_ty = np.sign(vel[1]) * 1/2 * m * vel[1] ** 2
        e_r = np.sign(rot_vel) * 1/2 * I * rot_vel ** 2
        
        print(f"\tTotal e = {abs(e_tx) + abs(e_r):.4}")
        print(f"\ttrans: {e_tx:.4}  rot: {e_r:.4}")
        
        oa = -np.arctan2(abs(vel[1]), abs(vel[0]))
        ca = (np.pi / 2) - abs(oa)
        a = np.pi - (2 * ca)
        area = ((a / (2 * np.pi)) * (r*r*np.pi)) - (1 / 2 * r * r * abs(np.sin(a)))
        k = ((r*r*np.pi) - 2*area)/(r*r*np.pi)
        print(f"\ta = {abs(oa) * (180/np.pi):.4}  k = {k:.4}")
        
        tr_y = e_ty * k * kty
        rt = e_r * kr #*(1-k)
        # tr_x = e_tx * ktx #*k
        
        e_ty -= tr_y
        e_r -= rt
        # e_tx -= tr_x
        
        if vel[0] > 0 and vel[1] > 0:
            e_r -= tr_y
            e_tx -= rt
        if vel[0] > 0 and vel[1] < 0:
            e_r -= tr_y
            e_tx += rt
        if vel[0] < 0 and vel[1] > 0:
            e_r += tr_y
            e_tx -= rt
        if vel[0] < 0 and vel[1] < 0:
            e_r += tr_y
            e_tx += rt
        
        e_r *= 0.8
        
        # print("\t",e_tx, e_r, tr, rt, e_r, e_tx)
        vel[1] = np.sign(e_ty)*(abs(e_ty * 2 / m) ** (1 / 2))
        vel[0] = np.sign(e_tx)*(abs(e_tx * 2 / m) ** (1 / 2))
        rot_vel = np.sign(e_r)*(abs(e_r * 2 /I) ** (1/2))
        vel[1] *= -1
        print(f"\tChanges it to: {vel}")
        print(f"\tTotal e = {abs(e_tx) + abs(e_r):.4}")
        print(f"\ttrans: {e_tx:.4}  rot: {e_r:.4}")
        
    t += dt
    angle += rot_vel * dt
    pos += vel * dt + 1/2 * acc * dt * dt
    vel += (accs[-1] + acc) / 2 * dt
    poss.append(pos.copy())
    vels.append(vel.copy())
    accs.append(acc.copy())
    angles.append(angle)
    times.append(t)
    spins.append(rot_vel)
    
fig, ax = plt.subplots(2, 3)
ax[0][0].plot(times, [p[0] for p in poss])
ax[0][0].set_xlabel("t")
ax[0][0].set_ylabel("x")
ax[0][1].plot(times, [p[1] for p in poss])
ax[0][1].set_xlabel("t")
ax[0][1].set_ylabel("y")
ax[0][2].plot([p[0] for p in poss], [p[1] for p in poss])
ax[0][2].set_xlabel("x")
ax[0][2].set_ylabel("y")

ax[1][0].plot(times, [p[0] for p in vels])
ax[1][0].set_xlabel("t")
ax[1][0].set_ylabel("x")
ax[1][1].plot(times, spins)
ax[1][1].set_xlabel("t")
ax[1][1].set_ylabel("y")
ax[1][2].plot([p[0] for p in poss], spins)
ax[1][2].set_xlabel("x")
ax[1][2].set_ylabel("y")
plt.show()
```
