```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fib(n, c1, c2):
    nums = [0, 1]
    for i in range(n):
        nums.append(c2 * nums[-2] + c1 * nums[-1])
    return nums
```

```python
num = 500
big = 0.7
small = 0.3
og = fib(num, big, small)
mod = fib(num, small, big)
```

```python
fig, ax = plt.subplots()
ax.plot(og, label=f"{big} + {small}")
ax.plot(mod, label=f"{small} + {big}")
plt.legend()
```

```python
print(og[-1])
print(mod[-1])
```
